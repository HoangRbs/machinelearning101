```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
np.random.seed(0)
points = 500
X = np.linspace(-3, 3, points)

# noises for standard sin(x) line to spread out the data
# apply for all 500 points
noises = np.random.uniform(-0.5, 0.5, points)

y = np.sin(X) + noises
```

```python
# ****** SOME NOTES ********

arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])

# only works with adding 2 NUMPY arrays
# not with regular python arrays
arr1 + arr2
```

```python
plt.scatter(X, y)
# plt.plot(X, predictions, 'ro')
plt.show()
```

```python
model = Sequential()
model.add(Dense(50, activation='sigmoid', input_dim=1)) # input x
model.add(Dense(30, activation='sigmoid')) # hidden layer
model.add(Dense(1)) # output y correspond to a single output 
```

```python
adam = Adam(lr=0.01)
model.compile(loss='mse', optimizer=adam)
model.summary()
```

```python
model.fit(X, y, epochs=50)
```

```python
predictions = model.predict(X)
plt.scatter(X, y) # scatter for showing points
plt.plot(X, predictions, 'ro') # draw line (model)
plt.show()
```
